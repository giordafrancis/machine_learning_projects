```python
# 3. Prepare Data
# a) Data Cleaning
# b) Feature Selection
# c) Data Transforms
```

```python
# a) Data Cleaning
cond_na = df.select_dtypes(np.object).isna().any(axis=1)
df = df[~cond_na]
```

```python
# c) Data transforms
df_obj = df.select_dtypes(np.object).copy().drop(columns = ['education', 'income_class'])
df_obj.describe().T
```

```python
df_obj['native_country_bin'] = np.where(df_obj['native_country'].str.strip() == 'United-States', 'native_ctr_usa', 'native_ctr_other')
df_obj['workclass_bin'] = np.where(df_obj['workclass'].str.strip() == 'Private', 'work_private', 'work_other')
```

```python
to_keep = df_obj['occupation'].value_counts()[df_obj['occupation'].value_counts().lt(3000)].index
df_obj.loc[df_obj['occupation'].isin(to_keep), 'occupation_bin'] = 'other'
df_obj['occupation_bin'] = df_obj['occupation_bin'].fillna(df_obj['occupation'])
```

```python
df_obj = df_obj.drop(columns = ['native_country', 'workclass', 'occupation'])
```

```python
df_obj_dummies = pd.get_dummies(df_obj, drop_first=True)
```

```python
df_obj_dummies.shape
```

```python
df_new = df.select_dtypes(np.number).join(df_obj_dummies).join(df['income_class'])
```

```python
# 4. Evaluate Algorithms
# a) Split-out validation dataset
# b) Test options and evaluation metric
# c) Spot Check Algorithms
# d) Compare Algorithms
```

```python
seed = 42
array = df_new.copy()
X = array.iloc[:, :-1]
y = array.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = seed)
```

```python
# for now baseline at the numeric options only
#X_train = X_train.select_dtypes(np.number)
```

```python
# baseline no data transform
models = []
models.append(['LogReg', LogisticRegression(solver='liblinear')])
models.append(['LDA', LinearDiscriminantAnalysis()])
models.append(['KNN', KNeighborsClassifier()])
models.append(['CART', DecisionTreeClassifier()])
models.append(['NB', GaussianNB()])
#models.append(['SVC', SVC(gamma='auto')]) # takes a long time retry once scales are normalised
```

```python
y_train_ = y_train.eq(' >50K').astype(int)
```

```python
num_folds = 10
# used accuarcy as les then 80% examples belong to majority class
scoring = 'accuracy'

results = {}
for name, model in models:
    #kfold = KFold(n_splits = num_folds, random_state=seed)
    kfold = StratifiedKFold(n_splits=num_folds, random_state=seed, shuffle=True)
    cv_results = cross_val_score(estimator=model, X=X_train, y=y_train_, scoring=scoring, cv=kfold)
    results[name] = cv_results
    msg =f" {name} mean:{cv_results.mean()} (+/-) {cv_results.std()}"
    print(msg)
```

```python
pd.DataFrame(results).plot(kind='box', title='Algorithm Comparison');
```

```python
# data scaled
pipelines = []
pipelines.append(('ScaledLR', Pipeline([('Scaler', StandardScaler()),('LR',
LogisticRegression(solver='lbfgs'))])))
pipelines.append(('ScaledLDA', Pipeline([('Scaler', StandardScaler()),('LDA',
LinearDiscriminantAnalysis())])))
pipelines.append(('ScaledKNN', Pipeline([('Scaler', StandardScaler()),('KNN',
KNeighborsClassifier())])))
pipelines.append(('ScaledCART', Pipeline([('Scaler', StandardScaler()),('CART',
DecisionTreeClassifier())])))
pipelines.append(('ScaledNB', Pipeline([('Scaler', StandardScaler()),('NB',
GaussianNB())])))
```

```python
scaled_results = {}
for name, pipeline in pipelines:
    #kfold = KFold(n_splits = num_folds, random_state=seed)
    kfold = StratifiedKFold(n_splits=num_folds, random_state=seed, shuffle=True)
    cv_results = cross_val_score(estimator= pipeline, X=X_train, y=y_train_,
                                scoring=scoring, cv=kfold)
    scaled_results[name] = cv_results
    msg =f" {name} mean:{cv_results.mean()} (+/-) {cv_results.std()}"
    print(msg)
```

```python
pd.DataFrame(scaled_results).plot(kind='box', title='Algorithm Comparison');
```

```python
estimators = [
    ('LogReg', LogisticRegression(solver='lbfgs', max_iter=100)),
    ('LDA', LinearDiscriminantAnalysis()),
    ('KNN', KNeighborsClassifier()),
    ('CART', RandomForestClassifier(n_estimators=10)),
    ('NB', GaussianNB())
]
k_best = SelectKBest(score_func=f_classif, k=15)
scaler = StandardScaler()
```

```python
scaled_kbest_results = {}
for name, estimator in estimators:
    kfold = KFold(n_splits = num_folds, random_state=seed)
    pipeline = make_pipeline(k_best, scaler, estimator)
    cv_results = cross_val_score(estimator= pipeline, X=X_train, y=y_train_,
                                scoring=scoring, cv=kfold)
    final_name = 'kbest_'+ name
    scaled_kbest_results[final_name] = cv_results
    msg =f" {final_name} mean:{cv_results.mean()} (+/-) {cv_results.std()}"
    print(msg)
```

```python
pd.DataFrame(scaled_kbest_results).plot(kind='box', title='Algorithm Comparison kbest',
                                     figsize=(12,8));
```

```python
# 5. Improve Accuracy
# a) Algorithm Tuning
# b) Ensembles - leave for now
```

```python
## set alphas for L2 search hyper param tunning 
pipe1 = make_pipeline(k_best, scaler, LogisticRegression(solver='lbfgs', max_iter=100))
pipe2 = make_pipeline(k_best, scaler, GaussianNB())

param_grid = {
    'selectkbest__k': np.arange(5,21,2),
    'logisticregression__C' : np.logspace(-4, 4, 10)
             }
#param_grid_nb = {'selectkbest__k': np.arange(5,21,2)}
#pipe.get_params()
```

```python
grid = GridSearchCV(estimator=pipe1, param_grid=param_grid, cv=5, verbose=True, scoring=scoring)
grid_results = grid.fit(X_train, y_train)
```

```python
grid_results.best_score_, grid_results.best_params_
```

```python
# save best params
C = grid_results.best_params_['logisticregression__C']
k = grid_results.best_params_['selectkbest__k']
```

```python
model = make_pipeline(SelectKBest(score_func=f_classif,
                                k=k), 
                     StandardScaler(),
                    LogisticRegression(C=C,
                                      solver='lbfgs',
                                      max_iter=100))
```

```python
model.fit(X_train, y_train)
predictions = model.predict(X_test)
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

### Conclusion

The model recall for the positive class >50k is circa 0.75 for the GNB classifier. I seems to handle the class imbalance better then Log regression.

In order to tune data to f1 score for minority class, the  desired minority output label in y_train vector needs to e labeled as `1`. If this is the objective for model tuning. 

Scoring with weighted f1 or roc_auc with lead to a better optmization overall but would miss more examples of the minority class.
